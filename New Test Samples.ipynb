## Docstrings

```
def add_nums(num1, num2):
    '''
    Add two numbers

    >>> add_nums(2,2)
    4
    '''
    return num1 + num2
```

```
help(add_nums)
```

```
add_nums(1,4)
```

```
def minus_nums(num1,num2):
    '''
    Subtract two numbers
    '''
    return num2-num1
```

```
num1 = 10
num2 = 5

print(add_nums(num1,num2))
print(minus_nums(num1,num2))
```
